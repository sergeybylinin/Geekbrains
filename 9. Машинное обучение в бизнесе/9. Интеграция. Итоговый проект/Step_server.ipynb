#### Сервис для обработки запросов на Flask

```python
from flask import Flask, request, jsonify
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score, roc_curve
import dill
dill._dill._reverse_typemap['ClassType'] = type
```

```python
# Загружаем обученные модели
with open('gbc_pipeline.dill', 'rb') as in_strm:
    model = dill.load(in_strm)
```

```python
# Обработчики и запуск Flask
app = Flask(__name__)
#run_with_ngrok(app)  # Start ngrok when app is run

@app.route('/predict', methods=['GET', 'POST'])
def predict():
    data = {"success": False}
    # ensure an image was properly uploaded to our endpoint
    if request.method == "POST":
        Date = ''
        Location = ''
        MinTemp = ''
        MaxTemp = ''
        Rainfall = ''
        Evaporation = ''
        Sunshine = ''
        WindGustDir = ''
        WindGustSpeed = ''
        WindDir9am = ''
        WindDir3pm = ''
        WindSpeed9am = ''
        WindSpeed3pm = ''
        Humidity9am = ''
        Humidity3pm = ''
        Pressure9am = ''
        Pressure3pm = ''
        Cloud9am = ''
        Cloud3pm = ''
        Temp9am = ''
        Temp3pm = ''
        RainToday = ''
        
        request_json = request.get_json()
        
    if request_json["Date"]: Date = request_json["Date"]
    if request_json["Location"]: Location = request_json["Location"]
    if request_json["MinTemp"]: MinTemp = request_json["MinTemp"]
    if request_json["MaxTemp"]: MaxTemp = request_json["MaxTemp"]
    if request_json["Rainfall"]: Rainfall = request_json["Rainfall"]
    if request_json["Evaporation"]: Evaporation = request_json["Evaporation"]
    if request_json["Sunshine"]: Sunshine = request_json["Sunshine"]
    if request_json["WindGustDir"]: WindGustDir = request_json["WindGustDir"]
    if request_json["WindGustSpeed"]: WindGustSpeed = request_json["WindGustSpeed"]
    if request_json["WindDir9am"]: WindDir9am = request_json["WindDir9am"]
    if request_json["WindDir3pm"]: WindDir3pm = request_json["WindDir3pm"]
    if request_json["WindSpeed9am"]: WindSpeed9am = request_json["WindSpeed9am"]
    if request_json["WindSpeed3pm"]: WindSpeed3pm = request_json["WindSpeed3pm"]
    if request_json["Humidity9am"]: Humidity9am = request_json["Humidity9am"]
    if request_json["Humidity3pm"]: Humidity3pm = request_json["Humidity3pm"]
    if request_json["Pressure9am"]: Pressure9am = request_json["Pressure9am"]
    if request_json["Pressure3pm"]: Pressure3pm = request_json["Pressure3pm"]
    if request_json["Cloud9am"]: Cloud9am = request_json["Cloud9am"]
    if request_json["Cloud3pm"]: Cloud3pm = request_json["Cloud3pm"]
    if request_json["Temp9am"]: Temp9am = request_json["Temp9am"]
    if request_json["Temp3pm"]: Temp3pm = request_json["Temp3pm"]
    if request_json["RainToday"]: RainToday = request_json["RainToday"]
        
    preds = model.predict_proba(pd.DataFrame({
        "Date": [Date],
        "Location": [Location],
        "MinTemp": [MinTemp],
        "MaxTemp": [MaxTemp],
        "Rainfall": [Rainfall],
        "Evaporation": [Evaporation],
        "Sunshine": [Sunshine],
        "WindGustDir": [WindGustDir],
        "WindGustSpeed": [WindGustSpeed],
        "WindDir9am": [WindDir9am],
        "WindDir3pm": [WindDir3pm],
        "WindSpeed9am": [WindSpeed9am],
        "WindSpeed3pm": [WindSpeed3pm],
        "Humidity9am": [Humidity9am],
        "Humidity3pm": [Humidity3pm],
        "Pressure9am": [Pressure9am],
        "Pressure3pm": [Pressure3pm],
        "Cloud9am": [Cloud9am],
        "Cloud3pm": [Cloud3pm],
        "Temp9am": [Temp9am],
        "Temp3pm": [Temp3pm],
        "RainToday": [RainToday]
    }))
    data["predictions"] = preds[:, 1][0]
    # indicate that the request was a success
    data["success"] = True
    
    # return the data dictionary as a JSON response
    return jsonify(data)


if __name__ == '__main__':
    app.run()
```
