#### 1. Если мы рассчитали 95% доверительный интервал для среднего значения, то какие из следующих утверждений являются верными?

Выберите несколько вариантов и попытайтесь объяснить свой выбор:

1) Если многократно повторять эксперимент, то 95 % выборочных средних значений будут принадлежать рассчитанному нами доверительному интервалу.

Верно. Считается, что выборочное среднее значение попадает в рассчитанный 95% доверительный интервал с вероятностью 95%. Значит, 95% выборочных средних значений будут попадать в рассчитанный нами доверительный интервал.

2) Мы можем быть на 95% уверены, что среднее значение в генеральной совокупности принадлежит рассчитанному доверительному интервалу.

Верно. Считается, что истинное значение принадлежит рассчитанному нами доверительному интервалу с вероятностью 95%.

3) Если многократно повторять эксперимент, для каждой выборки рассчитывать свой доверительный интервал, то в 95 % случаев истинное среднее будет находиться внутри доверительного интервала.

Верно.
Если мы рассчитываем свой доверительный интервал для каждой выборки, значит, наши выборки должны быть разного размера. Иначе 95% доверительный интервал должен быть одинаковым для всех выборок. Тем не менее, даже в этом случае для каждой выборки истинное среднее должно попадать в каждый из рассчитанных нами доверительных интервалов с вероятностью 95%. Значит, можно утверждать, что истинное среднее будет находиться внутри доверительного интервала в 95% случаев.

4) Среднее значение в генеральной совокупности точно превышает нижнюю границу 95% доверительного интервала.

Неверно. Есть небольшая вероятность, 2.5%, что среднее значение в генеральной совокупности не превысит нижнюю границу 95% доверительного интервала.

5) Среднее значение в генеральной совокупности точно принадлежит рассчитанному доверительному интервалу.

Неверно. Среднее значение в генеральной совокупности принадлежит рассчитанному доверительному интервалу с вероятностью 95%.

Верно. У нас достаточно оснований отклонить нулевую гипотезу при заданном 95% доверительном интервале и заданном уровнее значимости 0.05.

Неверно. При заданном 95% доверительном интервале эта вероятность равна 0.025 (уровень значимости 0.05 разделим на 2).

Верно. У нас достаточно оснований отклонить нулевую гипотезу при заданном 95% доверительном интервале и заданном уровнее значимости 0.05.

#### 6. Предположим, нулевой гипотезой вашего исследования являлось предположение, что конверсия в генеральной совокупности равняется 0.4. Вы получили p = 0,12 и не смогли отклонить нулевую гипотезу. Однако позже выяснилось, что конверсия в генеральной совокупности действиетльно равна 0.4. Как можно оценить результаты?

Какой вариант корректный по вашему мнению и почему?
1) Вы не совершали ни ошибку первого рода, ни ошибку второго рода.
2) Вы совершили ошибку первого рода
3) Вы совершили ошибку второго рода

Первый вариант. Мы не совершали ни ошибку первого рода, ни ошибку второго рода.
При ошибке первого рода отвергается верная нулевая гипотеза.
При ошибке второго рода принимается неверная нулевая гипотеза.
В нашем случае мы приняли верную нулевую гипотезу, так как конверсия в генеральной совокупности действительно равна 0.4.
Если мы не отвергли нулевую гипотезу, значит установленный нами уровень значимости был ниже 0.12, и выборочное среднее попало в доверительный интервал.
